# Now Let's get the Magnetic Field for a Pure Dipole

$B_{\hat{x}} = \frac{\mu}{4\pi \varepsilon _0}\frac{1}{c^2} \frac{3 XZ }{R^5}$

$B_{\hat{y}} = \frac{\mu}{4\pi \varepsilon _0}\frac{1}{c^2} \frac{3 YZ }{R^5}$

$B_{\hat{z}} = \frac{\mu}{4\pi \varepsilon _0}\frac{1}{c^2} \frac{3 Z Z }{R^5} - \frac{\mu}{4\pi \varepsilon _0}\frac{1}{c^2} \frac{1}{R^3} $

alt. def.: $B_{\hat{z}} = -\frac{\mu}{4\pi \varepsilon _0}\frac{1}{c^2} \left( \frac{1}{R^3} - \frac{3 Z^2 }{R^5}\right)$


```python
# Make the grid/"domain"
x0   = np.array([0.555, 0.555]) ## dipole location
C    = 10.0

r    = np.arange(-1, 2, 0.08)
Y, Z = np.meshgrid(r, r)
R    = np.array([Y, Z])

δR   = R - x0[:,None, None]
R    = np.linalg.norm(δR, axis=0)

By   = C * δR[0]*δR[1] / R**5
Bz   = C * δR[1]*δR[1] / R**5 - 1 / R**3

plt.streamplot(Y, Z, By, Bz)
plt.show()

plt.quiver(Y, Z, (By), (Bz))
plt.xlim([0.4,0.8])
plt.xlim([0.4,0.8])
plt.show()

# Make the plot
plt.pcolormesh(X, Y, div( np.array([By, Bz]) ), shading='auto')
plt.colorbar()
plt.show()
```

# Wire

\begin{align*}
    \mathcal{B}_{\hat{x}} &= -\frac{I}{2\pi\varepsilon_0} \frac{1}{c^2} \frac{Y}{ \left( \sqrt{X^2 + Y^2} \right)^{2} } \\
    \mathcal{B}_{\hat{y}} &= +\frac{I}{2\pi\varepsilon_0} \frac{1}{c^2} \frac{X}{ \left( \sqrt{X^2 + Y^2} \right)^{2} } \\
    \mathcal{B}_{\hat{z}} &= 0.
\end{align*}

## Curl & Divergence

```python
def div(v):
    """
    Given a Vector Field get a Scalar Field
    """
    divergence = np.zeros(list(v.shape)[1:], dtype=v.dtype)
    for i in range(len(v)):
        divergence += np.gradient(v[i], axis=i)
    return divergence

def curl(v):
    """
    Assumes 2D or 3D Vector Field
    """
    if len(v) == 2:
        ### This gets a Scalar Field
        curl    = np.zeros(list(v.shape)[1:], dtype=v.dtype)
        curl    = np.gradient(v[0], axis=1) - np.gradient(v[1], axis=0)
        return curl
    if len(v) == 3:
        curl    = np.zeros(v.shape, dtype=v.dtype)
        curl[0] = np.gradient(v[2], axis=1) - np.gradient(v[1], axis=2)
        curl[1] = np.gradient(v[0], axis=2) - np.gradient(v[2], axis=0)
        curl[2] = np.gradient(v[1], axis=0) - np.gradient(v[0], axis=1)
        return curl
    else:
        return None


v = np.asarray([Bx, By])

# Make the plot
plt.pcolormesh(X, Y, curl(v), shading='auto')
plt.colorbar()
plt.show()

# Make the plot
plt.pcolormesh(X, Y, div(v), shading='auto')
plt.colorbar()
plt.show()
```

# Magnetic Dipole field from Neutron (of some particle, just good units)

```python
def Solenoid3D(H,R,C,ds):
  B = np.zeros((3, ds, ds, ds))
  for i in range(ds-1): #one less so that we don't throw an error for indexing out of bounds
    dL = np.array(H[:,i+1] - H[:,i])
    δR   = R - H[:,i,None,None,None] #difference between the coordinates of a point on helix and the coordinates of every gridpoint, vector from helix to gridpoint
    Rn   = np.linalg.norm(δR, axis=0) #length, or magnitude, of the vector for each coordinate returned in the last step.
    B   +=   C * np.cross(dL, (δR/Rn), axis=0) / Rn**2
  return B
```

```python
mu_not = 1.257 * 10**-6
mag_mom = -9.6623647 * 10**-27 #magnetic moment for a neutron according to wikipedia, equal to area times current according to textbook
radius = 0.8 *10**-15 #mean square radius of a neutron, I'm going to take some liberties here.
current = mag_mom / (np.pi * radius**2)
C= current * mu_not / 4 / np.pi

ds = 50
txy = np.linspace(-3*radius,3*radius,ds)
t = np.linspace(0,6.29,ds)
X, Y, Z = np.meshgrid(txy,txy,txy)
R = np.array([X,Y,Z])
x = radius * np.cos(t)
z = 0*t
y = radius * np.sin(t)
circle_points = np.array([x,y,z])

ax = plt.figure().add_subplot(projection='3d')
ax.set(xlabel="X", ylabel="Y", zlabel="Z", title = 'Current Loop that will Model Magentic Field for a Neutron')
ax.plot3D(x, y, z,color='maroon')
plt.show()
```

As can be seen by the scatter plot as well as the slice of the magnitude array, values are nearly identical within the middle of the solenoid. This is due every point on the coil having a symeterical point lying on the opposite side. The vector components that are not parallel to the coil's length will cancel out when the vectors are all added, leaving only the parallel component. Each small segment of the coil can be treated as one source of magentic field. The size of this small segment is set by an argument ds when building the meshgrid. We can calculate the magnitude and direction of the small segment in refrence to every point in our grid using Biot-Savart law. Magnetic fields obey superposition so we can simply add all the given vectors for every piece along our coil to calculate the entire field.
