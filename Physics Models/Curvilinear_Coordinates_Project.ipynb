```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.neighbors import KDTree
X,Y = np.meshgrid(np.linspace(-4,4,30),np.linspace(-4,4,30))
a = 1
u = np.linspace(0.01,4,250,)
v  = np.linspace(0.01,2*np.pi,250,)
U,V = np.meshgrid(u,v)
Su = np.sinh(U)
Sv = np.sin(V)
Cv = np.cos(V)
Cu = np.cosh(U)
huv = 1/np.sqrt(Su**2+Sv**2)
uhat_x = huv *Cv*Su
uhat_y = huv *Cu*Sv
x = np.cosh(U) * np.cos(V) * a
y = np.sinh(U)*np.sin(V) * a

vhat_x = huv *-Sv*Cu
vhat_y = huv *Su*Cv
keys = ['x', 'y']
coordsyst = np.vstack([x.ravel(), y.ravel(),uhat_x.ravel(), uhat_y.ravel(),
                       vhat_x.ravel(), vhat_y.ravel()])
positionsXY = np.vstack([X.ravel(), Y.ravel()])
xydf= pd.DataFrame(coordsyst.T,columns=('x', 'y','uhatx','uhaty','vhatx',
                                        'vhaty')).reset_index()
XYdf= pd.DataFrame(positionsXY.T,columns=keys).reset_index()
kdt1 = KDTree(xydf[keys].values, leaf_size=5, metric='euclidean')
kdt2 = KDTree(XYdf[keys].values, leaf_size=5, metric='euclidean')
xydf['kNN'] = kdt2.query(xydf[keys].values, k=1, return_distance=False)[:,0]
XYdf['kNN'] = kdt1.query(XYdf[keys].values, k=1, return_distance=False)[:,0]
kNN1 = xydf.merge(XYdf[['index'] + keys], left_on='kNN', right_on='index',
                  suffixes=('_a', '_b'))
kNN1 = kNN1.drop_duplicates('index_b')
xyk = np.asarray((kNN1['x_b'],kNN1['y_b']))
uhatk = np.asarray((kNN1['uhatx'],kNN1['uhaty']))
vhatk = np.asarray((kNN1['vhatx'],kNN1['vhaty']))
fig,ax=plt.subplots(1,2,sharey=True,sharex=True,figsize=(15,7.5))
ax[0].quiver(xyk[0],xyk[1],uhatk[0],uhatk[1],label='u basis vector',
             color='blue',angles='xy', scale_units='xy',scale=4)
ax[1].quiver(xyk[0],xyk[1],vhatk[0],vhatk[1],label='v basis vector',
             color='red', angles='xy', scale_units='xy',scale=4)
plt.xlim(-3,3)
plt.ylim(-3,3)
fig.suptitle('Vector Fields of EC basis vectors in x-y plane',fontsize=12);
fig.text(0.51, 0.04, 'X', ha='center')
ax[0].set_title('u basis')
ax[1].set_title('v basis')
fig.text(0.08, 0.5, 'Y', va='center', rotation='vertical')
fig.savefig('ECbasisfields.png')
```
